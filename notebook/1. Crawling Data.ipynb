```python
import os
import shutil

# Path folder di Colab
source_folder = '/content/tweets-data/'

# Path folder di Google Drive tempat file akan disimpan
destination_folder = '/content/drive/My Drive/menyala 🔥🔥/dataset/new/2022 2023'

# Buat folder tujuan di Google Drive kalau belum ada
os.makedirs(destination_folder, exist_ok=True)

# Copy seluruh isi folder ke Google Drive
for filename in os.listdir(source_folder):
    source_file = os.path.join(source_folder, filename)
    destination_file = os.path.join(destination_folder, filename)
    if os.path.isfile(source_file):
        shutil.copy(source_file, destination_file)
        print(f'File "{filename}" berhasil di-upload ke Google Drive di {destination_file}.')

print(f'Semua file dari folder "{source_folder}" berhasil di-upload ke Google Drive di folder "{destination_folder}".')
```

```python
#Juni 2019-Agustus 2020
# Path ke folder CSV files
folder_path =  '/content/tweets-data/'

# List semua files di folder
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]

# List buat nyimpen data dari masing-masing CSV
dataframes = []

# Baca masing-masing CSV dan simpan ke list dataframes
for file in all_files:
    try:
        # Mencoba membaca file dengan delimiter koma (,)
        df = pd.read_csv(file)
        if df.empty:
            print(f"Skipping empty file: {file}")
            continue
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"Error reading file {file}: {e}")
        # Jika terjadi error, coba baca file dengan delimiter semicolon (;)
        try:
            df = pd.read_csv(file, delimiter=';')
            if df.empty:
                print(f"Skipping empty file: {file}")
                continue
            dataframes.append(df)
            print(f"Successfully read file {file} using semicolon delimiter.")
        except pd.errors.ParserError as e:
            print(f"Error reading file {file} with both delimiters: {e}")
            continue
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file}")
        continue

# Gabungin semua dataframes jadi satu
merged_df = pd.concat(dataframes, ignore_index=True)
```
